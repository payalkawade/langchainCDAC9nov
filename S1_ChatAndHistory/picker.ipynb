```python
import random
from langchain_groq import ChatGroq
from langchain_cohere import ChatCohere
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_core.chat_history import InMemoryChatMessageHistory

import os
import configparser

from langchain_groq import ChatGroq
from langchain_cohere import ChatCohere

from langchain_core.messages import HumanMessage, SystemMessage

config = configparser.ConfigParser()
config.read('../config.ini')
groq = config['groq']
cohere = config['cohere']

os.environ['GROQ_API_KEY'] = groq.get('GROQ_API_KEY')
os.environ['COHERE_API_KEY'] = cohere.get('COHERE_API_KEY')

# Step 1: Create `actor_picker` function
def actor_picker():
    actors = ["Robert Downey", "Denzel Washington", "Tom Cruise", "Kevin Spacey"]
    return random.choice(actors)

# Step 2: Create `location_picker` function
def location_picker():
    locations = ["Switzerland", "US", "Turkey", "London"]
    return random.choice(locations)

# Step 3: Create `theme_picker` function
def theme_picker():
    themes = ["Adventure", "Romance", "Mystery", "Science Fiction"]
    return random.choice(themes)

# Step 4: Configure and use Cohere Model
def generate_story_with_cohere(actor, location, theme):
    model = ChatCohere(model="command-r-plus")
    messages = [
        SystemMessage(content="You are a creative story generator."),
        HumanMessage(content=f"Create a story with the actor {actor}, location {location}, and theme {theme}.")
    ]
    return model.invoke(messages).content

# Step 5: Configure and use Groq Model
def generate_story_with_groq(actor, location, theme):
    model = ChatGroq(model="llama3-8b-8192")
    messages = [
        SystemMessage(content="You are a creative story generator."),
        HumanMessage(content=f"Create a story with the actor {actor}, location {location}, and theme {theme}.")
    ]
    return model.invoke(messages).content

# Step 6: Add InMemoryChatMessageHistory
def customize_story_with_history(initial_story):
    store = InMemoryChatMessageHistory()
    store.add_message(SystemMessage(content="Here is the initial story:"))
    store.add_message(HumanMessage(content=initial_story))
    
    user_customization = "Change the story to include a hero who helps solve the mystery."
    store.add_message(HumanMessage(content=user_customization))
    
    model = ChatGroq(model="llama3-8b-8192")
    messages = store.messages + [
        SystemMessage(content="Modify the story based on the user request.")
    ]
    response = model.invoke(messages)
    return response.content

# Execution
if __name__ == "__main__":
    actor = actor_picker()
    location = location_picker()
    theme = theme_picker()

    print("### Selected Elements ###")
    print(f"Actor: {actor}")
    print(f"Location: {location}")
    print(f"Theme: {theme}")
    print("---")

    # Generate story using Cohere
    cohere_story = generate_story_with_cohere(actor, location, theme)
    print("### Story from Cohere Model ###")
    print(cohere_story)
    print("---")

    # Generate story using Groq
    groq_story = generate_story_with_groq(actor, location, theme)
    print("### Story from Groq Model ###")
    print(groq_story)
    print("---")

    # Customize story
    customized_story = customize_story_with_history(groq_story)
    print("### Customized Story ###")
    print(customized_story)
```
